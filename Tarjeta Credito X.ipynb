```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
from functools import reduce
```

```python
df = pd.read_excel("C:/Users/adria/OneDrive/Escritorio/Repaso Python Anlisis de datos/Tarjeta_Credito.xlsx")
df
```

```python
# Nuevas COLUMNAS creadas
```

```python
df["Dia_Fecha"] = df["Fecha_Compra"].dt.day # Obtener dias de la fecha

#Columna Periodo_Compra
condition = [
    df["Dia_Fecha"] >= 21,
    df["Dia_Fecha"] >=11
]
choice = ("Despues del dia 21","Entre el dia 11 y 20")
df["Periodo_Compra"] = np.select(condition, choice,default="Entre el dia 1 y 10")

#Columna Clasificacion_Satisfaccion_Cliente
condition = [
    df["Satisfaccion_Cliente"] >= 4,
    df["Satisfaccion_Cliente"] == 3
] 
choice = ("Alta","Media")
df["Clasificacion_Satisfaccion_Cliente"] = np.select(condition,choice,default="Baja")

#Clasificacion_Tipo_Compra
condition = [
    df["Monto_Pesos"] >= 150000,
    df["Monto_Pesos"] >= 80000,
    df["Monto_Pesos"] >= 35000
] 
choice = ("Premiun","Alto", "Estandard")
df["Clasificacion_Tipo_Compra"] = np.select(condition,choice,default="Bajo")
df

# Plan_Cuotas
condition = [
    df["Cuotas"] == 1,
    df["Cuotas"] == 3,
    df["Cuotas"] == 6,
]
choice = ("Plan 1 cuota","Plan 3 cuotas","Plan 6 cuotas")
df["Plan_Cuotas"] = np.select(condition,choice,default="Plan 12 coutas")
df
```

```python
## - Analisis de CLIENTES - Se dividio el analisis en tres partes (intensidad, frecuencia, montos)
```

```python
## 1- INTENSIDAD DE COMPRA
```

```python
#Agrupar por cliente y fecha y devolver el tamaño de esa combinacion, por ejemplo 1, 2 , 3 compras por dia y cliente
intensidad = df.groupby(['Cliente','Fecha_Compra']).size().reset_index(name='Compras_por_dia')
analisis_compras = intensidad.groupby("Cliente").agg(
    compras_Xdia_promedio = ("Compras_por_dia","mean"),
    Cantidad_Compras = ("Compras_por_dia","sum")
).round(2).sort_values(by="Cantidad_Compras",ascending=False)
analisis_compras["%Compras"] = ((analisis_compras["Cantidad_Compras"] / analisis_compras["Cantidad_Compras"].sum())*100).round(2).astype(str) + "%"
analisis_compras
```

```python
## 2 - FRECUENCIA DE COMPRA
```

```python
df_unico = df.drop_duplicates(subset=['Cliente','Fecha_Compra']).sort_values(['Cliente','Fecha_Compra'])
df_unico["Fecha_Compra_Anterior"] = df_unico.groupby("Cliente")["Fecha_Compra"].shift(1)
df_unico["Dias_Entre_Compras"] = (df_unico["Fecha_Compra"] - df_unico["Fecha_Compra_Anterior"]).dt.days
analisis_dias = df_unico.groupby("Cliente").agg(
    Frecuencia_Compra_Dias = ("Dias_Entre_Compras","mean"),
    Fecha_Ultima_Compra = ("Fecha_Compra","max")
).round(2).sort_values(by="Frecuencia_Compra_Dias")
analisis_dias["Fecha_de_Cierre"] = df_unico["Fecha_Cierre_Ciclo"].iloc[0]
analisis_dias["Dias_Desde_Ultimo_Consumo"] = (analisis_dias["Fecha_de_Cierre"] - analisis_dias["Fecha_Ultima_Compra"]).dt.days
analisis_dias["Desvio_del_Promedio"] = analisis_dias["Dias_Desde_Ultimo_Consumo"] - analisis_dias["Frecuencia_Compra_Dias"]
condition = [
    analisis_dias["Desvio_del_Promedio"] > 0,
     analisis_dias["Desvio_del_Promedio"] <=  0
]
choice = ("dias por encima del promedio","dias por debajo del promedio")
analisis_dias["Detalle_Desvio"] = np.select(condition, choice,default="otro")
analisis_dias
```

```python
## 3 - ANALISIS FINANCIERO
```

```python
analisis_financiero = df.groupby("Cliente").agg(
    Ticket_Promedio = ("Monto_Pesos","mean"),
    Monto = ("Monto_Pesos","sum"),
    Satisfaccion_Cliente = ("Satisfaccion_Cliente","mean")
).round(2).sort_values(by="Monto", ascending= False)

analisis_financiero["%Monto"] = ((analisis_financiero["Monto"] / analisis_financiero["Monto"].sum())*100).round(2)
analisis_financiero["%Acumulado_Monto"] = (analisis_financiero["%Monto"]).cumsum().round(2)

analisis_financiero["%Monto"] = analisis_financiero["%Monto"].astype(str) + "%"
analisis_financiero["Ticket_Promedio"] = analisis_financiero["Ticket_Promedio"].apply(lambda x: f"${x:,.2f}".replace(",","X").replace(".",",").replace("X","."))
analisis_financiero["Monto"] = analisis_financiero["Monto"].apply(lambda x: f"${x:,.2f}".replace(",","X").replace(".",",").replace("X","."))

condition = [
    analisis_financiero["%Acumulado_Monto"] <= 85,
    analisis_financiero["%Acumulado_Monto"] <=95 
]
choice = ("A","B")
analisis_financiero["Pareto_ClienteABC"] = np.select(condition,choice,default="C")

analisis_financiero["%Acumulado_Monto"] = analisis_financiero["%Acumulado_Monto"].astype(str) + "%"

analisis_financiero
```

```python
## 4- UNION DE LOS 3 HALLAZGOS
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)
dfs_analisis_clientes = [analisis_compras,analisis_dias,analisis_financiero]
df_consolidado_analisis_clientes = reduce(
    lambda left, right: pd.merge(left, right, on="Cliente", how="left"), 
    dfs_analisis_clientes)
df_consolidado_analisis_clientes = df_consolidado_analisis_clientes.sort_values(by="Monto", ascending = False)
df_consolidado_analisis_clientes
```

```python
## ANALISIS POR PERIODO DE COMPRA
```

```python
analisis_periodo_compras = df.groupby("Periodo_Compra").agg(
    Cantidad_Compras = ("Id_Transaccion","count"),
    Ticket_Promedio = ("Monto_Pesos","mean"),
    Recaudacion = ("Monto_Pesos","sum"),
    Medio_Pago_Mas_Utilizado = ("Canal_Pago",lambda x: x.mode()),
    Tipo_Gasto_Frecuente = ("Clasificacion_Tipo_Compra",lambda x: x.mode()),
    Plan_Cuota_Mas_Usado = ("Plan_Cuotas", lambda x: x.mode())
).round(2).sort_index().sort_values(by="Cantidad_Compras", ascending = False )

analisis_periodo_compras["%Recaudacion"] = ((analisis_periodo_compras["Recaudacion"] / analisis_periodo_compras["Recaudacion"].sum()) 
                                           * 100).round(2).astype(str) + "%"
analisis_periodo_compras["%Compras"] = ((analisis_periodo_compras["Cantidad_Compras"] / analisis_periodo_compras["Cantidad_Compras"].sum()) 
                                        *100).round(2).astype(str) + "%"

analisis_periodo_compras["Ticket_Promedio"] = analisis_periodo_compras["Ticket_Promedio"].apply(lambda x: f"${x:,.2f}".replace(",", "X").replace(".", ",").replace("X", "."))
analisis_periodo_compras["Recaudacion"] = analisis_periodo_compras["Recaudacion"].apply(lambda x: f"${x:,.2f}".replace(",", "X").replace(".", ",").replace("X", "."))

Orden = [
"Cantidad_Compras",
"%Compras",
"Tipo_Gasto_Frecuente",
"Ticket_Promedio",
"Recaudacion",
"%Recaudacion",
"Medio_Pago_Mas_Utilizado",
"Plan_Cuota_Mas_Usado"
]

analisis_periodo_compras = analisis_periodo_compras [Orden]
analisis_periodo_compras
```

```python
## ANALISIS POR PRODUCTO
```

```python
analisis_producto = df.groupby("Producto").agg(
    Ventas = ("Id_Transaccion","count"),
    Ticket_Promedio = ("Monto_Pesos","mean"),
    Recaudacion = ("Monto_Pesos","sum"),
    Medio_Pago_Mas_Usado = ("Canal_Pago",lambda x: x.mode()),
    Tipo_Gasto_Frecuente = ("Clasificacion_Tipo_Compra",lambda x: x.mode()),
    Periodo_Con_Mas_Compras = ("Periodo_Compra",lambda x: x.mode()),
    Plan_Cuota_Mas_Usado = ("Plan_Cuotas", lambda x: x.mode())
    
).round(2).sort_values(by = "Recaudacion", ascending = False )

analisis_producto["%Recaudacion"] = ((analisis_producto["Recaudacion"] / analisis_producto["Recaudacion"].sum()) * 100).round(2).astype(str) + "%"
analisis_producto["%Ventas"] = ((analisis_producto["Ventas"] / analisis_producto["Ventas"].sum()) * 100).round(2).astype(str) + "%"
analisis_producto["Ticket_Promedio"] = analisis_producto["Ticket_Promedio"].apply(lambda x: f"${x:,.2f}".replace(",","X").replace(".",",").replace("X","."))
analisis_producto["Recaudacion"] = analisis_producto["Recaudacion"].apply(lambda x: f"${x:,.2f}".replace(",","X").replace(".",",").replace("X","."))

Orden = [
"Ventas",
"%Ventas",
"Ticket_Promedio",
"Recaudacion",
"%Recaudacion",
"Periodo_Con_Mas_Compras",
"Medio_Pago_Mas_Usado",
"Plan_Cuota_Mas_Usado",
"Tipo_Gasto_Frecuente"
]
analisis_producto = analisis_producto[Orden]
analisis_producto
```

```python
## ANALISIS POR CANAL DE PAGO
```
